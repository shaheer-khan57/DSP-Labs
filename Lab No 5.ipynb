```python
import numpy as np
import matplotlib.pyplot as plt

# --- Z^-1 domain coefficients ---
X = np.array([1, 2, 1])   # X(z) = 1 + 2 z^-1 + 1 z^-2
H1 = np.array([1, -2])    # H1(z) = 1 - 2 z^-1
```

```python
# --- Y1(z) = X(z) * H1(z)
Y1 = np.polymul(X, H1)    # equivalent to convolution in time
```

```python
print("Y1(z) coefficients:", Y1)
```

```python
# --- Plotting ---
plt.figure(figsize=(10,4))

plt.subplot(1,2,1)
plt.stem(range(len(Y1)), Y1)
plt.title("y1[n] = X(z) * H1(z)")
plt.xlabel("n")
plt.ylabel("Amplitude")
plt.grid(True)
```

```python
# Truncate to N samples for numeric convolution
N = 3
H2 = np.array([2**n for n in range(N)])
print("h2[n] =", H2)
```

```python
Y2 = np.polymul(Y1,H2)
```

```python
print("Y2(z) coefficients:", Y2)
```

```python
# --- Only take the first 3 samples for plotting ---
Y1_plot = Y1[:3]
Y2_plot = Y2[:3]
```

```python
# Plot Y2
plt.subplot(1,2,2)
plt.stem(range(len(Y2_plot)), Y2_plot)
plt.title("y2[n] (first 3 samples)")
plt.xlabel("n")
plt.ylabel("Amplitude")
plt.grid(True)

plt.tight_layout()
plt.show()
```
