```python
# Task #02: Read, play, and change sampling frequency of an audio file

import librosa
import librosa.display
import sounddevice as sd
import matplotlib.pyplot as plt

# Step 1: Load the audio file
audio_path = "D:\\بیانات\\Audios\\Un...H.W.T.mp3"

# sr=None → keeps the original sampling rate
y, sr = librosa.load(audio_path, sr=None)

print(f"Original Sampling Rate: {sr} Hz")
print(f"Audio Duration: {len(y)/sr:.2f} seconds")

```

```python


# Step 3: Plot the original waveform
plt.figure(figsize=(10, 4))
librosa.display.waveshow(y, sr=sr)
plt.title(f"Original Audio Waveform (Sampling rate = {sr} Hz)")
plt.xlabel("Time (s)")
plt.ylabel("Amplitude")
plt.grid(True)
plt.show()

```

```python
# Step 4: Change sampling frequency (resample)
new_sr = 8000  # New sampling rate (for example, 8 kHz)
y_resampled = librosa.resample(y, orig_sr=sr, target_sr=new_sr)

print(f"🔁 Resampled to: {new_sr} Hz")


# Step 6: Plot the resampled waveform
plt.figure(figsize=(10, 4))
librosa.display.waveshow(y_resampled, sr=new_sr)
plt.title(f"Resampled Audio Waveform (Sampling rate = {new_sr} Hz)")
plt.xlabel("Time (s)")
plt.ylabel("Amplitude")
plt.grid(True)
plt.show()
```

```python
import sounddevice as sd
```

```python
# Step 5: Play the resampled audio
print("🎧 Playing resampled audio...")
sd.play(y_resampled, new_sr)
sd.wait()
```
