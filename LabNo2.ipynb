```python
import numpy as np
import matplotlib.pyplot as plt

# Signal parameters
f = 50          # frequency in Hz
T = 1/f         # period
fs = 1000       # sampling frequency (Hz) - must be much higher than f
t = np.arange(0, 2*T, 1/fs)  # time vector (2 cycles)

# Generate the signal
signal = np.sin(2 * np.pi * f * t)

# Plot the signal
plt.figure(figsize=(8, 4))
plt.plot(t, signal, label="50 Hz Sine Wave")
plt.xlabel("Time (seconds)")
plt.ylabel("Amplitude")
plt.title("50 Hz Signal")
plt.grid(True)
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Common sampling rate (must be high enough for both)
fs = 10000  # 10 kHz is fine because it’s > 2 * 500 Hz (Nyquist rule)
t = np.arange(0, 0.04, 1/fs)  # 40 ms duration (enough to see a few cycles)

# Signal 1: 50 Hz
f1 = 50
signal1 = np.sin(2 * np.pi * f1 * t)

# Signal 2: 500 Hz
f2 = 500
signal2 = np.sin(2 * np.pi * f2 * t)

# Combined signal
combined_signal = signal1 + signal2
```

```python
# Plot combined signal
plt.subplot(3, 1, 3)
plt.plot(t, combined_signal)
plt.title("Combined 50 Hz + 500 Hz Signal")
plt.xlabel("Time (seconds)")
plt.ylabel("Amplitude")
plt.grid(True)

plt.tight_layout()
plt.show()

```
